```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\91859\Desktop\ARYAN\ARYAN JUPYTER\sampledata\API_SP.POP.TOTL_DS2_en_csv_v2_1584446.csv"
data = pd.read_csv(file_path, skiprows=3)

```

```python
year = '2021'
selected_data = data[['Country Name', year]]

# Drop rows with  mmissing values
selected_data = selected_data.dropna()

#Plotting the bar chart
plt.figure(figsize=(20, 15))

plt.bar(selected_data['Country Name'], selected_data[year])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Population Distribution in {year}')
plt.xticks(rotation=90)

plt.show()
```
